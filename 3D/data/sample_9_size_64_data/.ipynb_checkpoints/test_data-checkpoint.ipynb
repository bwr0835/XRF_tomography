```python
import numpy as np
import torch as tc
import h5py
import os

import matplotlib.pyplot as plt
import matplotlib 
matplotlib.rcParams['pdf.fonttype'] = 'truetype'
fontProperties = {'family': 'sans-serif', 'sans-serif': ['Helvetica'], 'weight': 'normal', 'size': 12}
plt.rc('font', **fontProperties)
from matplotlib import gridspec
from mpl_toolkits.axes_grid1 import make_axes_locatable
import matplotlib.ticker as mtick
```

```python
data_path = "nElements_1"
f_XRF_data = "XRF_sample9"
f_XRT_data = "XRT_sample9"
theta_st = 0
theta_end = 360
n_theta = 200
theta_ls = - np.linspace(theta_st, theta_end, n_theta+1)[:-1]
```

## XRF data tranformation

```python
XRF_data = np.array([np.load(os.path.join(data_path, f_XRF_data)+'_{}'.format(i) + '.npy').astype(np.float32) for i in range(n_theta)])
XRF_data = np.reshape(XRF_data, (n_theta, 2, 64, 64))
print(XRF_data.shape)
XRF_data = np.transpose(XRF_data, (1,0,2,3))
print(XRF_data.shape)

elements_array = np.array(["Si", "Si_L"], dtype='S')
print(elements_array.dtype)
```

```python
with h5py.File("./nElements_1/test9_xrf", "w") as f:
    grp = f.create_group("exchange")
    data = grp.create_dataset("data", shape = (2, 200, 64, 64), dtype="f4",)
    elements = grp.create_dataset("elements", shape = (2,), dtype='S5')
    theta = grp.create_dataset("theta", shape = (n_theta,), dtype = "f4")
    
    data[...] = XRF_data
    elements[...] = elements_array
    theta[...] = theta_ls
```

```python
with h5py.File("./nElements_1/test9_xrf", "r") as f:
    data = f["exchange/data"][...]
    elements = f["exchange/elements"][...]
    
XRF_line_idx = 0
y_idx = 15
print(data[XRF_line_idx, :, y_idx, :].shape)

fig, ax = plt.subplots(figsize=(4,8), nrows=1, ncols=1)
sinogram = ax.imshow(data[XRF_line_idx, :, y_idx, :], cmap='jet', aspect=0.5, extent=[0, 64, 200, 0], origin='upper')                                                                                                  
divider = make_axes_locatable(ax)
cax = divider.append_axes('right', size='5%', pad='2%', aspect=31.3)
cbr = fig.colorbar(sinogram, cax=cax, orientation='vertical')
cax.xaxis.set_ticks_position('top')
cbr.ax.tick_params(labelsize=16) 
```

## XRT data tranformation

```python
XRT_data = np.array([np.load(os.path.join(data_path, f_XRT_data)+'_{}'.format(i) + '.npy').astype(np.float32) for i in range(n_theta)])
XRT_data = np.reshape(XRT_data, (n_theta, 1, 64, 64))
XRT_data = np.transpose(XRT_data, (1,0,2,3))
print(XRT_data.shape)

ds_ic = XRT_data
us_ic = np.full(XRT_data.shape, 1.0E7)
abs_ic = ds_ic / us_ic

scaler_data = np.concatenate((np.zeros(XRT_data.shape), us_ic, ds_ic, abs_ic), axis=0)
print(scaler_data.shape)
print(np.max(XRT_data))
print(np.min(XRT_data))
print(abs_ic.shape)
```

```python
plt.imshow(abs_ic[0, :, 22, :], cmap='jet', aspect=0.5, vmin=0.9, vmax= 1)
print(np.argwhere(abs_ic == np.min(abs_ic)))
```

```python
plt.imshow(ds_ic[0, :, 30, :], cmap='jet', aspect=0.5, vmin=0.0, vmax= 1.0E7)
```

```python
elements_array = np.array(["empty", "us_ic", "ds_ic", "abs_ic"], dtype='S')
```

```python
with h5py.File("./nElements_1/test9_xrt", "w") as f:
    grp = f.create_group("exchange")
    data = grp.create_dataset("data", shape = (4, 200, 64, 64), dtype="f4")
    elements = grp.create_dataset("elements", shape = (4,), dtype="S12")
    theta = grp.create_dataset("theta", shape = (n_theta,), dtype = "f4")
    
    data[...] = scaler_data
    elements[...] = elements_array
    theta[...] = theta_ls
```
